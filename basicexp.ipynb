### Some of the operations below are expectedly performed only on numerical attributes. As 'Sex' attribute is converted into numeric though it is categorical in nature, it is also subject to these operations. However, 'BMIcat' attribute is not numerical, thus, not exposed.

## General descriptive statistics of data

```python
print(data.describe())
```

## Minimum value of attributes

```python
print(data.min())
```

## Maximum value of attributes

```python
print(data.max())
```

## Minimum and maximum value of specific attribute

```python
print("min: {}, max: {}".format(data.DGestimate.min(), data.DGestimate.max()))
```

## Mean of attributes

```python
print(data.mean())
```

## Standard deviation of attributes

```python
print(data.std)
```

## Mode of attributes

```python
print(data.mode())

# For "DGerror" attribute, there are three modes while for others, there is only one.
```

## Median of attributes

```python
print(data.median())
```

## Variance of attributes

```python
print(data.var())
```

## Mean, variance, standard deviation, mode and median of specific attribute

```python
print("mean: {}, variance: {}, standard deviation: {}, mode: {}, median: {}".format(round(data.SBestimate.mean(), 2), round(data.SBestimate.var(), 2), round(data.SBestimate.std(), 2), data.SBestimate.mode().iloc[0], data.SBestimate.median()))
```

## Pairwise covariance of attributes

```python
print(round(data.cov(), 2))
```

## Covariance of two specific attributes:

```python
print(round(data.BMIquant.cov(data.Age), 2))
```

## Pairwise correlation of attributes

```python
print(round(data.corr(), 2))
```

## Correlation between two specific attributes

```python
print(round(data.BMIquant.corr(data.Age), 2))
```
